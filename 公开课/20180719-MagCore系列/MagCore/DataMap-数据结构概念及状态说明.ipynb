## 游戏(Game)
---

- Id: 游戏的Id
- Map: 游戏所使用的地图名称
- State: 当前游戏的状态
- Players: 参加当前游戏的玩家**数组**
- Cells: 当前游戏的单元状态**二维数组**

## 玩家(Player)
---

- Id: 玩家Id
- Name: 玩家昵称
- Token: 玩家安全令牌(暂保留)
- Energy: 玩家能量(暂保留)
- Color: 玩家颜色
- State: 玩家状态
- Index: 玩家标识号
- Bases:玩家基地坐标**数组**

## 单元(Cell)
---

- X: 单元的x坐标
- Y: 单元的y坐标
- Type: 单元的类型
- State: 单元的状态
- Owner: 单元所属的玩家标识号

## 状态
---

### 游戏状态(GameState)
---

- 0: 等待加入
- 1: 游戏中
- 2: 结束
- 3: 已销毁

### 地图单元类型(CellType)
---

- 0: 空
- 1: 正常单元
- 2: 基地

### 单元状态(CellState)
---

- 0: 无归属
- 1: 被攻占中
- 2: 被占领

### 玩家状态(PlayerState)
---

- 0: 空闲
- 1: 已加入游戏
- 2: 被击败

### 玩家颜色(PlayerColor)
---
- 0: 香蕉
- 1: 樱桃
- 2: 葡萄
- 3: 菜瓜
- 4: 柠檬
- 5: 桑葚
- 6: 生梨
- 7: 菠萝
- 8: 萝卜
- 9: 西瓜
