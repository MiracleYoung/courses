```python

lst = list(range(1000000))
s = set(range(1000000))
```

```python
%%timeit
-1 in lst
```

```python
def _slice(lst, start=0, stop=0, step=1):
    if stop == 0:
        stop = len(lst)
        
    current = start
    ret = []
    
    while current >= start and current < stop:
        try:
            ret.append(lst[current])
        except IndexError:
            return ret
        current += step
    return ret
```

```python
_slice(lst, 0, 10, -2)
```

```python
lst[0:10:-2]
```

```python
def _slice(lst, start=0, stop=0, step=1):
    if stop == 0:
        stop = len(lst)
    
    if step < 0 :
        _start = start * -1
        _stop = stop * -1
    else:
        _start = start
        _stop = stop
        
    current = start
    ret = []
    
    while current >= _start and current < _stop:
        try:
            ret.append(lst[current])
        except IndexError:
            return ret
        current += step
    return ret
```

```python
def _slice(lst, start=0, stop=0, step=1):
    
    ret = []
    current = start
    
    if step > 0:
        while current < stop:
            try:
                ret.append(lst[current])
            except IndexError:
                pass
            current += step
    if step < 0:
        while current > stop:
            try:
                ret.append(lst[current])
            except IndexError:
                pass
            current += step
    return ret
```
