### for

```
开始
for 元素 in 可迭代对象：
    操作
结束
```

````python
r = range(0, 10)

for i in r:
    print(i)
````

````python
list(range(0, 10, 2))
````

````python
r = range(0, 10)


for i in r:
    print(i)
    if i == 4:
        break
        print('break')
````

````python
r = range(0, 10)


for i in r:
    for j in [1, 2]:
        print(i)
        if i == 4:
            break
            print('break')
            
# break 跳出的是 当前层
````

````python
r = range(0, 10)


for i in r:

    print(i)
    if i == 4:
        continue
    print('break')
            
# break 跳出的是 当前层, 中段 当前循环
# continue 跳过本次 循环
````

````python
for i in range(10):
    pass
else:
    print('ok')
````

````python
for i in range(10):
    if i == 3:
        continue
else:
    print('ok')
````

````python
for i in range(10):
    if i == 3:
        break
else:
    print('ok')
````

**for...else 中的else，要在 for循环正确结束后 才被执行**

2. 有1，2，3，4个数字，能组成多少个互不相同且不重复数字的三位数

123, 124, 

122, 334

````python
l = [1, 2, 3, 4]
number = 0
for x in l:
    for y in l:
        if x == y:
            continue
        else:
            for z in l:
                if y == z or z == x:
                    continue
                else:
                    number += 1
                    print(100*x + 10*y + z)
````
